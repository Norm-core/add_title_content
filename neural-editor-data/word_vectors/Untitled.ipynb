```python
vectors = []
with open('vector.txt') as f:
    for line in f.readlines():
        vectors.append(line.strip())
len(vectors)
```

```python
dicts = []
with open('dict.txt') as f:
    for line in f.readlines():
        dicts.append(line.strip())
len(dicts)
```

```python
dicts[:5]
```

```python
tags = {}
for i in range(len(dicts)):
    tags[dicts[i].split(' ')[0]] = dicts[i].split(' ')[2]
len(tags)
```

```python
tags['中国']
```

```python
word_cixing = {}
for i in range(len(words)):
    try:
        word_cixing[words[i]] = tags[words[i]]
    except:
        word_cixing[words[i]] = 'none'
```

```python
word_cixing['中国'], word_cixing['。']
```

```python
list(word_cixing.values())[:5]
```

```python
cixings = list(word_cixing.values())
len(cixings)
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# define example
data = cixings
values = array(data)
print(values[:5])
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded[:5])
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded[:5])
# invert first example
inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
print(inverted)
```

```python
cixing_vec = {}
for i in range(len(cixings)):
    cixing_vec[cixings[i]] = list(onehot_encoded[i])
len(cixing_vec)
```

```python
len(cixing_vec['none']), cixing_vec['none']
```

```python
word_cixing_vec = {}
for word in list(word_cixing.keys()):
    word_cixing_vec[word] = cixing_vec[word_cixing[word]]
```

```python
len(word_cixing_vec)
```

```python
print(word_cixing_vec['中国'])
print(len(word_cixing_vec['中国']))
```

```python
cixing_dic = {}
with open('vector_cixing.txt') as f:
    f.readline()
    for line in f.readlines():
        cixing_dic[line.strip().split(' ')[0]] = line.strip().split(' ')[1:]
len(cixing_dic)
```

```python
len(cixing_dic['中国']), cixing_dic['中国']
```

```python
ner_ls = []
with open('ner.txt') as f:
    for line in f.readlines():
        ner_ls.append(line.strip())
len(ner_ls)
```

```python
ner_ls[:5]
```

```python
cixing_dic['中国'] + ['1.0']
```

```python
for word in list(cixing_dic.keys()):
    if word in ner_ls:
        cixing_dic[word] += ['1.0']
    else:
        cixing_dic[word] += ['0.0']
```

```python
len(cixing_dic['中国'])
```

```python
with open('vector_add.txt', 'w') as f:
    f.write('87390 157\n')
    for word in list(cixing_dic.keys()):
        f.write(word + ' ' + ' '.join(cixing_dic[word]) + '\n')
```
